# Automatic Way

```python
from sklearn.preprocessing import LabelEncoder
clean_species_id = LabelEncoder()
clean_sex = LabelEncoder()
df['clean_sex'] = clean_sex.fit_transform(df['sex'].astype(str)) 
df['clean_species_id'] = clean_species_id.fit_transform(df['species_id'].astype(str))
```

```python
df.columns
```

```python
df=df[['record_id','year','plot_id','hindfoot_length', 'weight','clean_species_id','clean_sex']]
df.head()
```
